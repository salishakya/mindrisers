```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme()
tips = sns.load_dataset("tips")
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    col="time",
    hue="smoker",
    style="smoker",
    size="size",
)
```

```python
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots,
    kind="line",
    x="time",
    y="firing_rate",
    col="align",
    hue="choice",
    style="choice",
    size="coherence",
    facet_kws=dict(sharex=False),
)
```

```python
sns.lmplot(
    data=tips,
    x="total_bill",
    y="tip",
    col="time",
    hue="smoker",
)
```

```python
# histogram
sns.displot(
    data=tips,
    x="total_bill",
    col="time",
)
```

```python
sns.displot(data=tips, x="total_bill", col="time", kde=True)
```

```python
# cat plot is to cehck outliers
sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
    kind="swarm",
    hue="smoker",
)
```

```python
sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
    kind="violin",
    hue="smoker",
)
```

```python
sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
    kind="bar",
    hue="smoker",
)
```
